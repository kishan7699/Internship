# QUESTION2

```python
urls=[]
for page in range(0,3):
    try:
        page_urls=driver.find_elements_by_xpath('//a[@class="a-link-normal s-link-style a-text-normal"]')
        for url in page_urls:
            url = url.get_attribute('href')
            if url[0:4]=='http':
                urls.append(url)
        print("Product urls of page {} has been scraped.".format(page+1))    
        
        nxt_button=driver.find_elements_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        try:
            driver.get(nxt_button[0].get_attribute('href'))
        except:
            driver.get(nxt_button[1].get_attribute('href'))
            
            
        
    except NoSuchElementException as e:
        print("No Such Exception" ,e)
        next_page=nxt_button.get_attribute('href')
        driver.get(next_page)
        
                                                
    
```

```python
len(urls)
```

```python
PRODUCT=[]
BRANDNAME=[]
PRICE=[]
RETEXC=[]
EXPDEL=[]
AVAIL=[]
for j in urls:
    driver.get(j)
    time.sleep(2)
    try:
        product=driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]')
        PRODUCT.append(product.text.replace("/n","newline"))
    except:
        PRODUCT.append('-')
    
    try:
        brand=driver.find_element_by_xpath('//td[@class="a-span9"]/span')
        BRANDNAME.append(brand.text.replace("/n","newline"))
    except:
        BRANDNAME.append('-')
        
    try:
        price=driver.find_element_by_xpath('//span[@class="a-price aok-align-center priceToPay"]/span[2]')
        PRICE.append(price.text.replace("\n","."))
    except:
        PRICE.append('-')
        
            
    try:
        Return=driver.find_element_by_xpath('//div[@class="a-section a-spacing-none icon-content"]')
        RETEXC.append(Return.text.replace("/n","newline"))
    except:
        RETEXC.append('-')
    
    try:
        expdel=driver.find_element_by_xpath('//div[@class="a-section a-spacing-mini"]/b')
        EXPDEL.append(expdel.text.replace("/n","newline"))
    except:
        EXPDEL.append('-')
        
    try:
        avail=driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        AVAIL.append(avail.text.replace("/n","newline"))
    except:
        AVAIL.append('-')        
    
```

```python
len(PRODUCT),len(BRANDNAME),len(PRICE),len(RETEXC),len(EXPDEL),len(AVAIL),len(urls)
```

```python
Guitar=pd.DataFrame({})
Guitar['PRODUCT']=PRODUCT
Guitar['BRANDNAME']=BRANDNAME
Guitar['PRICE']=PRICE
Guitar['RETURN/EXCHANGE']=RETEXC
Guitar['EXPECTED DELIVERY']=EXPDEL
Guitar['AVAILABILITY']=AVAIL
Guitar['URL']=urls


Guitar
```

```python
urls10=[]
for i in driver.find_elements_by_xpath('//h2[@class="title title-6"]/a'):
    urls10.append(i.get_attribute('href'))
           
```

```python
urls10
```

```python

Hostelname=[]
Distance=[]
Rating=[]
TotalReview=[]
OverallReview=[]
Private=[]
Dorms=[]

Property=[]
for H in urls10:
    driver.get(H)
    time.sleep(2)
    try:
        hostel=driver.find_element_by_xpath('//div[@class="title-2"]')
        Hostelname.append(hostel.text.replace("/n","newline"))
    except:
        Hostelname.append('-')
    try:
        distance=driver.find_element_by_xpath('//div[@class="body-3"]')
        Distance.append(distance.text.replace("/n","newline"))
    except:
        Distance.append('-')
    try:
        rating=driver.find_element_by_xpath('//div[@class="score orange big"]')
        Rating.append(rating.text.replace("\n","."))
    except:
        Rating.append('-')
    try:
        total=driver.find_element_by_xpath('//div[@class="rating-factors rating-factors big"]')
        TotalReview.append(total.text.replace("\n",","))
    except:
        TotalReview.append('-')
    try:
        overall=driver.find_element_by_xpath('//ul[@class="rating-breakdown rating-grid"]')
        OverallReview.append(overall.text.replace("\n"," "))
    except:
        OverallReview.append('-')
    try:
        private=driver.find_element_by_xpath('//div[@class="price"]')
        Private.append(private.text.replace("/n","newline"))
    except:
        Private.append('-')
    try:
        dorms=driver.find_element_by_xpath('//div[@class="rate-list"]')
        Dorms.append(dorms.text.replace("\n"," "))
    except:
        Dorms.append('-')
    driver.find_element_by_xpath('//a[@class="toggle-content"]').click()
    try:
        propert=driver.find_element_by_xpath('(//div[@class="collapsible"]')
        Property.append(propert.text.replace("/n","newline"))
    except:
        Property.append('-')
        
    
```

```python
Hostel=pd.DataFrame({})
Hostel['Hostelname']=Hostelname
Hostel['Distance']=Distance
Hostel['Rating']=Rating
Hostel['TotalReview']=TotalReview
Hostel['OverallReview']=OverallReview
Hostel['Private']=Private
Hostel['Dorms']=Dorms
Hostel['Property']=Property

Hostel
```
